## Prepare df_i and df_h to be merged

```python
# Drop Size_Adjusted_Gross_Income column from df_i as df_i2

df_i2 = df_i.drop('Size_Adjusted_Gross_Income', axis=1)
```

```python
# Show df_i2

df_i2.head()
```

```python
# Add average income per individual column in df_i2 called average_income

df_i2['average_income'] = df_i2['Total_Income'] / df_i2['Num_Individuals']
```

```python
# Show df_i2

df_i2.head()
```

```python
# Multiply average_income by 1000 to show accurate dollar amount

df_i2['average_income'] = df_i2['average_income'] * 1000
```

```python
# Show df_i2

df_i2.head()
```

```python
# drop column Num_Individuals in df_i2

df_i2 = df_i2.drop('Num_Individuals', axis = 1)
```

```python
# drop column Total_Income in df_i2

df_i2 = df_i2.drop('Total_Income', axis=1)
```

```python
# show df_i2 head

df_i2.head()
```

```python
# Group average_income by ZIP_Code in df_i2 as df_iz using .mean

df_iz = df_i2.groupby('ZIP_Code')['average_income'].mean()
```

```python
# Show df_iz head

df_iz.head()
```

```python
# turn df_iz_2 into a dataframe

df_iz_2 = pd.DataFrame(df_i2.groupby('ZIP_Code')['average_income'].mean(), columns=['average_income'])
print(df_iz_2)
```

```python
# Show df_iz_2

df_iz_2.head()
```

```python
# Change ZIP_Code from an index to a colum value called zip_code

df_iz_3 = df_iz_2.reset_index()
```

```python
# rename column ZIP_Code to zip_code in df_iz_3

df_iz_3.rename(columns = {'ZIP_Code': 'zip_code'}, inplace = True)
```

```python
# rename column State to state in df_i2

df_i2.rename(columns = {'State': 'state'}, inplace = True)
```

```python
# Show df_i2

df_i2.head()
```

```python
# drop column Unnamed in df_i2

df_iz = df_i2.drop('Unnamed: 0', axis=1)
```

```python
# Drop column Unnamed in df_h

df_h = df_h.drop('Unnamed: 0', axis=1)
```

```python
# Show df_i2

df_i2.head()
```

```python
# Show df_iz_2

df_iz_3.head()
```

```python
# Show df_h

df_h.head()
```

# Merge dataframes

```python
# Merge df_h with df_iz_2 by zip_code as df_merged

df_merged = pd.merge(df_h, df_iz_3, on='zip_code', how='inner')
```

```python
# Check df_merged head

df_merged.head()
```

```python
# Add unique identification column using index

df_merged_2 = df_merged.reset_index()
```

```python
# Check df_merged head

df_merged_2.head()
```

```python
# Rename column index to id in df_merged_2

df_merged_2.rename(columns = {'index': 'id'}, inplace = True)
```
